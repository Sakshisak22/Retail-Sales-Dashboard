```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Month_Name')['Total Amount'].sum()

plt.figure()
monthly_sales.plot(kind='bar')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Total Amount'].sum()

plt.figure()
yearly_sales.plot()
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_categories = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(5)

plt.figure()
top_categories.plot(kind='bar')
plt.title("Top 5 Product Categories")
plt.ylabel("Total Sales")
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum()

plt.figure()
gender_sales.plot(kind='bar')
plt.title("Sales by Gender")
plt.ylabel("Total Sales")
plt.show()
```

```python
bins = [18, 25, 35, 45, 60]
labels = ['18-25', '26-35', '36-45', '46-60']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_sales = df.groupby('Age Group')['Total Amount'].sum()

plt.figure()
age_sales.plot(kind='bar')
plt.title("Sales by Age Group")
plt.ylabel("Total Sales")
plt.show()
```

```python
monthly_data = df.resample('M', on='Date')['Total Amount'].sum().reset_index()
monthly_data.head()
```

```python
monthly_data['Month_Number'] = range(len(monthly_data))
```

```python
from sklearn.linear_model import LinearRegression

X = monthly_data[['Month_Number']]
y = monthly_data['Total Amount']

model = LinearRegression()
model.fit(X, y)
```

```python
future_months = 6
future_X = [[i] for i in range(len(monthly_data), len(monthly_data) + future_months)]

predictions = model.predict(future_X)

print("Next 6 Months Sales Prediction:")
print(predictions)
```

```python
plt.figure()
plt.plot(monthly_data['Month_Number'], y)
plt.plot(range(len(monthly_data), len(monthly_data) + future_months), predictions)
plt.title("Sales Forecast")
plt.xlabel("Month Number")
plt.ylabel("Total Sales")
plt.show()
```
